## Matplotlib 특징

* 파이썬의 대표적인 과학 계산용 그래프 라이브러리
* 선 그래프, 히스토그램, 산점도 등의 고품질 그래프 제공
* 저수준 api를 사용한 다양한 시각화 기능 제공
* 다양한 운영체제와 그래픽 백엔드에서 동작


```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
mpl.__version__
```

```python
plt.style.use(['seaborn-notebook'])
```

## 라인 플롯(Line Plot)

* 플롯(plot)은 그림(figure)와 축(axes)으로 구성
* `plt.Figure`: 축과 그래픽, 텍스트, 레이블을 표시하는 모든 객체를 포함하는 컨테이너
* `plt.Axes`: 눈금과 레이블이 있는 테두리 박스로 시각화를 형성하는 플롯 요소 포함

```python
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure()
plt.plot([0, 0.2, 0.4, 0.6, 0.8, 1.] * 5); # 텍스트 정보를 없애려고 ; 를 써줌
```

```python
x = np.arange(0, 10, 0.01)

fig = plt.figure()
plt.plot(x, np.sin(x));
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));
```

```python
plt.plot(np.random.randn(50).cumsum());
```

```python
cols = list('ABC')
colors = [np.random.rand(3), np.random.rand(3), np.random.rand(3)]
df = pd.DataFrame(columns=cols, data=[[0, 1, 2], [0, 1, 2], [0, 1, 3]])
for j,i in enumerate(cols):
    df[i].plot(c=colors[j], label=i)
plt.legend()
```

```python
plt.plot([np.random.choice(['A','B','C']) for i in range(50)],np.random.randn(50).cumsum());
```

```python
plt.plot(np.random.randn(50).cumsum())
plt.xticks(np.arange(50),[np.random.choice(['A','B','C']) for i in range(50)]);
```

## x축, y축 그리기

```python
plt.xlim(0,3)
plt.ylim(0,3)
plt.axvline(x=1, color = 'k', ls=':')  # draw x =0 axes
plt.axhline(y=1, color = 'k', ls=':')   # draw y =0 axes
plt.show();
```

## 라인 스타일(Line Style)


|문자|스타일|
|:---|:----
|`-`|`solid`|
|`--`|`dashed`|
|`-.`|`dashdot`|
|`:`|`dotted`|

```python
plt.plot(np.random.randn(50).cumsum(), linestyle='solid')
plt.plot(np.random.randn(50).cumsum(), ls='dashed')
plt.plot(np.random.randn(50).cumsum(), linestyle='dashdot')
plt.plot(np.random.randn(50).cumsum(), ls='dotted')
```

```python
plt.plot(np.random.randn(50).cumsum(), linestyle='-')
plt.plot(np.random.randn(50).cumsum(), ls='--')
plt.plot(np.random.randn(50).cumsum(), linestyle='-.')
plt.plot(np.random.randn(50).cumsum(), ls=':')
```

## 색상 스타일(Color Style)


|문자|색상|
|:-:|:-|
|`'b'`|blue|
|`'e'`|green|
|`'r'`|red|
|`'c'`|сyan|
|`'m'`|magenta|
|`'y'`|yellow|
|`'k'`|black|
|`'w'`|wihte|


image.png

```python
plt.plot(np.random.randn(50).cumsum(), color='g')
plt.plot(np.random.randn(50).cumsum(), c='#FF43FF')
plt.plot(np.random.randn(50).cumsum(), color=(1., 0.9, 0.8))
plt.plot(np.random.randn(50).cumsum(), c='darkblue')
plt.plot(np.random.randn(50).cumsum(), color=np.random.rand(3))
```

```python
plt.plot(np.random.randn(50).cumsum(), c='skyblue')
plt.plot(np.random.randn(50).cumsum(), color='dodgerblue')
plt.plot(np.random.randn(50).cumsum(), c='royalblue')
plt.plot(np.random.randn(50).cumsum(), color='navy');
```

```python
# 한번에 라인과 색상을 같이
plt.plot(np.random.randn(50).cumsum(), 'b-')
plt.plot(np.random.randn(50).cumsum(), 'g--')
plt.plot(np.random.randn(50).cumsum(), 'c-.')
plt.plot(np.random.randn(50).cumsum(), 'm:')
```
