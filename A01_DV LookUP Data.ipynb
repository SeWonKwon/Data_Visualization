## First thing frist

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use(['seaborn-whitegrid'])
```

### Distribution, outlier, quantile, std, mean, median

* Distribution : `sns.pairplot`
* Outlier : `boxplot`
* Quantile : `boxplot`
* Std : `boxplot`, `histplot`
* median : `boxplot`
* skewness : `histplot`
* mean : `barplot`

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
sns.relplot(x='sepal_length', y='sepal_width',data=iris)
```

```python
sns.pairplot(iris)
```

```python
iris.boxplot()
```

```python
iris.hist();
```

```python
plt.hist(iris.iloc[:,:-1])
```

```python
d = dict(histtype='stepfilled', alpha=0.3, bins=50)
for i in iris.columns[:-1]:
    plt.hist(iris[i],**d)
```

```python
sns.boxplot(data=iris);
```

```python
sns.histplot(data=iris);
```

```python
sns.histplot(data=iris, kde=True);
```

```python
sns.barplot(iris.columns[:-1],iris.mean());
```

```python
sns.barplot(iris.mean(), iris.columns[:-1]);
```
